```python
paper_smiles = "CNC(C)CC1=CC=C2C(=C1)OCO2"
```

```python
def return_canonical_deb(smiles):
    return Chem.MolToSmiles(Chem.MolFromSmiles(smiles))
    
```

```python
#better use Chem.c
#one canonical is not alway another canonical: https://github.com/rdkit/rdkit/issues/2747

def return_canonical(smiles):
    m = Chem.MolFromSmiles(smiles)
    m.UpdatePropertyCache(strict=False)
    Chem.SanitizeMol(m,Chem.SanitizeFlags.SANITIZE_FINDRADICALS|Chem.SanitizeFlags.SANITIZE_KEKULIZE|Chem.SanitizeFlags.SANITIZE_SETAROMATICITY|Chem.SanitizeFlags.SANITIZE_SETCONJUGATION|Chem.SanitizeFlags.SANITIZE_SETHYBRIDIZATION|Chem.SanitizeFlags.SANITIZE_SYMMRINGS,catchErrors=True)
    return Chem.MolToSmiles(Chem.MolFromSmiles(smiles))
```

```python
fail_smiles = "C C ( C ) ( O ) C 1 C C N C C 1 . C C N ( C ( C ) C ) C ( C ) C . C N ( C ) C ( O n 1 n n c 2 c c c n c 2 1 ) = [N+] ( C ) C . Cl C Cl . C n 1 c ( C ( = O ) O ) n c 2 c ( N 3 C C O C C 3 ) n c ( - n 3 c ( C 4 C C 4 ) n c 4 c c c c c 4 3 ) n c 2 1 . F [P-] ( F ) ( F ) ( F ) ( F ) F".replace(" ","")
```

```python
CanonSmiles(fail_smiles) == return_canonical_deb(fail_smiles)
```

```python
another = "C C C C P ( C C C C ) C C C C . C O C ( = O ) C c 1 c n ( C ) c 2 c c ( O ) c c c 1 2 . C c 1 c c c c c 1 . C c 1 n n ( - c 2 c c c ( C ( F ) ( F ) F ) c c 2 ) c c 1 C ( C ) C O".replace(" ","")
```

```python
CanonSmiles(another) == another
```
